```python
import roulette_table_class as table 
wheel = table.RouletteTable()
for i in range(11):
    wheel.spin()
print(wheel.pastSpins)
balance = 0
```

```python
bet_color = None
last_bet_amount = 0
bet_amount = 0
def place_initial_bet(prev):
    #pass previous color to determine next bet
    # Parameters: prev - the previously spun color
    # initial bet is set to 5
    global bet_amount, last_bet_amount
    bet_amount = 5
    last_bet_amount = bet_amount
    if prev == 'red':
        color = 'black'
    elif prev == 'black':
        color = 'red'
    print(f"Placing initial bet of {bet_amount} on {color}")
    return color

def double_bet_amount(last):
    #double previous bet and update the last_bet_amount and bet_amount variables
    #Parameter: last - previously bet amount
    global bet_amount, last_bet_amount
    bet_amount = last*2
    if bet_amount>100:
        bet_amount = 5 # if bet exceeds 100, reset to 5
    last_bet_amount = bet_amount
```

```python

def simulate(times):
    """Simulates the strategy I wanted to test. Wait until the same color is spun twice in a row, the bet the opposite color. Print statements at the end to summarize results of the simulation.
    Parameter: times - how many times you want to spin the wheel"""
    max_bet = 0
    max_down = 0
    max_up = 0
    green_lands = 0
    global balance, bet_color, bet_amount, last_bet_amount
    for b in range(times):
        print("Current bet amount:",bet_amount)
        if bet_amount > max_bet:
            max_bet = bet_amount
            print("New max bet:",max_bet)
        wheel.spin()
        print("Spin",b+1,":",wheel.get_last_spin())
        if wheel.get_last_spin() == 'green':
            green_lands+=1
        if bet_amount != 0:
            if wheel.get_last_spin() == bet_color:
                balance += bet_amount
                bet_amount = 0
            else:
                balance -= bet_amount
                double_bet_amount(last_bet_amount)
        else:
            if wheel.pastSpins[-1] == wheel.pastSpins[-2] and wheel.pastSpins[-1] != 'green':
                bet_color = place_initial_bet(wheel.pastSpins[-1])
        print(f"Balance: {balance}")
        if balance > max_up:
            max_up = balance
        if balance < max_down:
            max_down = balance
    print("Final balance:", balance)
    print("Max bet placed:", max_bet)
    print("Max up:", max_up)
    print("Max down:", max_down)
    print("Green lands:", green_lands)
    
```

```python
balance = 0
simulate(240)
```
